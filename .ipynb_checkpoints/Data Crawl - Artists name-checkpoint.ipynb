```python
import pandas as pd
from selenium import webdriver
import csv
from string import ascii_lowercase
```

```python
chrome_path = "/Users/sheetalpandrekar/Documents/Jupyter Notebooks/ATnC/Spire/chromedriver"
driver=webdriver.Chrome(chrome_path)
```

# Crawl artist names and hrefs a-z

```python
for ch in ascii_lowercase:
    
    if ch == "a":
        continue

    link = "https://www.azlyrics.com/"+ch+".html"
    driver.get(link)

    with open("Data/artists_name_"+ch+".csv", "a", newline='') as fp:
        wr = csv.writer(fp, dialect='excel')

        elm = driver.find_elements_by_xpath("/html/body/div[2]/div/div[1]/a")
        for e in elm:
            artist_name = e.text
            artist_href = e.get_attribute('href')
            wr.writerow([artist_name,artist_href])


        elm = driver.find_elements_by_xpath("/html/body/div[2]/div/div[2]/a")
        for e in elm:
            artist_name = e.text
            artist_href = e.get_attribute('href')
            wr.writerow([artist_name,artist_href])
```

# Crawl the page 19.html for artist names and href

```python
link = "https://www.azlyrics.com/19.html"
driver.get(link)

with open("Data/artists_name_19.csv", "a", newline='') as fp:
    wr = csv.writer(fp, dialect='excel')

    elm = driver.find_elements_by_xpath("/html/body/div[2]/div/div[1]/a")
    for e in elm:
        artist_name = e.text
        artist_href = e.get_attribute('href')
        wr.writerow([artist_name,artist_href])


    elm = driver.find_elements_by_xpath("/html/body/div[2]/div/div[2]/a")
    for e in elm:
        artist_name = e.text
        artist_href = e.get_attribute('href')
        wr.writerow([artist_name,artist_href])
```

# Get all unique links

```python
hrefs = list()

for ch in ascii_lowercase:
    print(ch)
    f = pd.read_csv("Data/artists_name_"+ch+".csv", header=None)
    f.columns = ['artist_name','href']
    hrefs.extend(f['href'].tolist())

print(hrefs)
```

```python
print(len(hrefs))
```

```python
u_hrefs = list(set(hrefs))
print(len(u_hrefs))
```

```python
s = pd.Series(u_hrefs)
s.to_csv("Data/links.csv", index=False, header=False)
```
