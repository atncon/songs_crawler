# Get the songs for each artist

We have a list of artist_names. We call the genius.search_artist() function by passing the artist name as an argument. <br>
We can specify the max_songs parameter to limit the number of songs requested from the API. <br>

Set get_full_info=False to speed up the response. <br>

Set allow_name_change = False to prevent the API from changing the queried name. If  allow_name_change = True, it will look for the nearest matching artist name

save_lyrics() function saves the fetched data in 'Lyrics_{artist_name}.json'

```python
artist_names = ['Aaron Fresh', 'A92', 'Cynthia Erivo']

for a_name in artist_names:
    
    artist = genius.search_artist(a_name, get_full_info=False, allow_name_change = False)
    #artist = genius.search_artist(a_name, max_songs=20, get_full_info=False, allow_name_change = False)
    
    os.getcwd()
    artist.save_lyrics()
```

# Read the .json response

We can read the 'Lyrics_{artist_name}.json' file to get the Artist ID, Artist Name, Song ID, Song Title, Song Lyrics

```python
import json

for a_name in artist_names:
    a_name = a_name.replace(" ","")
    
    f = open("Lyrics_"+a_name+".json") 
    data = json.load(f) 
    f.close()

    print("Artist ID: %d" %data['id'])
    print("Artist name: %s" %data['name'])

    print("Song ID: %d" %data['songs'][0]['id'])
    print("Song title: %s" %data['songs'][0]['title'])
    print("Song lyrics: %s" %data['songs'][0]['lyrics'])
    print()
```
